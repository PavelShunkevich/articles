```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

data = np.random.normal(loc=0, scale=1, size=2000)

fig, (ax_box, ax_dist) = plt.subplots(nrows=2, sharex=True,  figsize=(12, 8))  #     

box_plot_data = ax_box.boxplot(data, vert=False,
                            medianprops=dict(color="black"))
ax_box.set(yticks=[], xlabel="")

ax_dist.hist(data, bins=30, density=True, color='skyblue', edgecolor='black')
ax_dist.set_ylabel("Density")

stats_data = {}
for key, val in box_plot_data.items():
    stats_data[key] = [item.get_xdata() for item in val]
q1 = stats_data['boxes'][0][0]
q3 = stats_data['boxes'][0][2]
median = stats_data['medians'][0][0]
whisker_lower = stats_data['whiskers'][0][1]
whisker_upper = stats_data['whiskers'][1][1]


ax_dist.axvline(q1, color='r', linestyle='--', label='Q1', linewidth=3)
ax_dist.axvline(median, color='g', linestyle='-', label='Median', linewidth=3)
ax_dist.axvline(q3, color='r', linestyle='--', label='Q3', linewidth=3)
ax_dist.axvline(whisker_lower, color='b', linestyle=':', label='Whisker Lower', linewidth=3)
ax_dist.axvline(whisker_upper, color='b', linestyle=':', label='Whisker Upper', linewidth=3)

# Annotations
arrowprops = dict(arrowstyle='->', color='black', linewidth=1.5)
text_offset_x = 0.05 * (ax_box.get_xlim()[1] - ax_box.get_xlim()[0])
text_offset_y = 0.1
fontsize = 13

ax_box.annotate('1. Median', xy=(median, 0.93), xytext=(median + 0, 0.93 - text_offset_y),
                    arrowprops=arrowprops, ha='center', va='top', fontsize=fontsize)
ax_box.annotate('2. Q1', xy=(q1, 1.07), xytext=(q1 - text_offset_x, 1.07 + text_offset_y),
                    arrowprops=arrowprops, ha='right', va='bottom', fontsize=fontsize)
ax_box.annotate('3. Q3', xy=(q3, 1.07), xytext=(q3 + text_offset_x, 1.07 + text_offset_y),
                    arrowprops=arrowprops, ha='left', va='bottom', fontsize=fontsize)
ax_box.annotate('5. Outliers', xy=(whisker_lower-0.3, 0.95), xytext=(whisker_lower-0.3,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='center', va='top', fontsize=fontsize)
ax_box.annotate('5. Outliers', xy=(whisker_upper+0.3, 0.95), xytext=(whisker_upper+0.3,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='center', va='top', fontsize=fontsize)
ax_box.annotate('6.1. Whisker Lower', xy=(whisker_lower, 0.95), xytext=(whisker_lower + text_offset_x,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='left', va='top', fontsize=fontsize)
ax_box.annotate('6.2. Whisker Upper', xy=(whisker_upper, 0.95), xytext=(whisker_upper - text_offset_x,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='right', va='top', fontsize=fontsize)

ax_dist.set(xticks=[], xlabel="") 
plt.tight_layout()
plt.savefig("boxplot_norm_annotated.png")  
plt.show()
```

```python
import numpy as np

#  
data = np.array([10, 12, 15, 18, 20, 22, 25, 30, 40, 70])

#  IQR
Q1 = np.quantile(data, 0.25)
Q3 = np.quantile(data, 0.75)
IQR = Q3 - Q1

print(" : ", data)  
print(f"IQR ( ): {IQR}")
```

```python
import numpy as np

#  
data = np.array([10, 12, 15, 18, 20, 22, 25, 30, 40, 70])

#  Q1, Q3  IQR
Q1 = np.quantile(data, 0.25)
Q3 = np.quantile(data, 0.75)
IQR = Q3 - Q1

#    
upper_outlier_threshold = Q3 + 1.5 * IQR
lower_outlier_threshold = Q1 - 1.5 * IQR

#    ( ,    )
upper_whisker = data[data <= upper_outlier_threshold].max()

#    ( ,    )
lower_whisker = data[data >= lower_outlier_threshold].min()

print(" : ", data)
print(f" : {upper_whisker}")
print(f" : {lower_whisker}")
```

```python
import numpy as np

#  
data = np.array([10, 12, 15, 18, 20, 22, 25, 30, 40, 70])

#  Q1, Q3  IQR
Q1 = np.quantile(data, 0.25)
Q3 = np.quantile(data, 0.75)
IQR = Q3 - Q1

#    
upper_outlier_threshold = Q3 + 1.5 * IQR
lower_outlier_threshold = Q1 - 1.5 * IQR

#  
outliers = data[(data < lower_outlier_threshold) | (data > upper_outlier_threshold)]


print(" : ", data)
print(f"  : {upper_outlier_threshold}")
print(f"  : {lower_outlier_threshold}")
print(f": {outliers}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

#      
data = {'Toyota': np.random.normal(loc=150, scale=25, size=24),
        'Honda': np.random.normal(loc=130, scale=20, size=24),
        'Ford': np.random.normal(loc=100, scale=30, size=24),
        'BMW': np.random.normal(loc=80, scale=15, size=24),
        'Tesla': np.random.normal(loc=90, scale=20, size=24)}
#  
data['Ford'][0] = 20.1
data['Ford'][1] = 30.312
data['Tesla'][1] = 150.112

#  DataFrame
data = pd.DataFrame(data)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))  #      
plt.boxplot(data, labels=data.columns)
# sns.boxplot(data=data)  #  sns.boxplot   
plt.title("     ( 24 )")
plt.ylabel("  ")
plt.grid(True)
plt.show()
```
