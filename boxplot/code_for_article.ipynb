```python
## Import libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

np.random.seed(42)
```

```python
#Screensaver picture

data1 = np.random.normal(loc=50, scale=15, size=100)
data2 = np.random.normal(loc=70, scale=10, size=100)
data3 = np.random.normal(loc=60, scale=20, size=100)
data = [data1, data2, data3]

palette = ["#a8dadc", "#457b9d", "#77ACF1"]

plt.figure(figsize=(12, 8))
sns.boxplot(data=data, palette=palette)
plt.savefig("boxplot_pastel.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=2000)

fig, (ax_box, ax_dist) = plt.subplots(nrows=2, sharex=True,  figsize=(12, 8))  #     

box_plot_data = ax_box.boxplot(data, vert=False,
                            medianprops=dict(color="black"))
ax_box.set(yticks=[], xlabel="")

ax_dist.hist(data, bins=30, density=True, color='skyblue', edgecolor='black')
ax_dist.set_ylabel("Density")

stats_data = {}
for key, val in box_plot_data.items():
    stats_data[key] = [item.get_xdata() for item in val]
q1 = stats_data['boxes'][0][0]
q3 = stats_data['boxes'][0][2]
median = stats_data['medians'][0][0]
whisker_lower = stats_data['whiskers'][0][1]
whisker_upper = stats_data['whiskers'][1][1]


ax_dist.axvline(q1, color='r', linestyle='--', label='Q1')
ax_dist.axvline(median, color='g', linestyle='-', label='Median')
ax_dist.axvline(q3, color='r', linestyle='--', label='Q3')
ax_dist.axvline(whisker_lower, color='b', linestyle=':', label='Whisker Lower')
ax_dist.axvline(whisker_upper, color='b', linestyle=':', label='Whisker Upper')

# Annotations
arrowprops = dict(arrowstyle='->', color='black', linewidth=1.5)
text_offset_x = 0.05 * (ax_box.get_xlim()[1] - ax_box.get_xlim()[0])
text_offset_y = 0.1

ax_box.annotate('1. Q1', xy=(q1, 1.07), xytext=(q1 - text_offset_x, 1.07 + text_offset_y),
                    arrowprops=arrowprops, ha='right', va='bottom', fontsize=10)
ax_box.annotate('2. Median', xy=(median, 0.93), xytext=(median + 0, 0.93 - text_offset_y),
                    arrowprops=arrowprops, ha='center', va='top', fontsize=10)
ax_box.annotate('3. Q3', xy=(q3, 1.07), xytext=(q3 + text_offset_x, 1.07 + text_offset_y),
                    arrowprops=arrowprops, ha='left', va='bottom', fontsize=10)
ax_box.annotate('4. Whisker Lower', xy=(whisker_lower, 0.95), xytext=(whisker_lower + text_offset_x,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='left', va='top', fontsize=10)
ax_box.annotate('5. Whisker Upper', xy=(whisker_upper, 0.95), xytext=(whisker_upper - text_offset_x,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='right', va='top', fontsize=10)
ax_box.annotate('6. Outliers', xy=(whisker_lower-0.3, 0.95), xytext=(whisker_lower-0.3,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='center', va='top', fontsize=10)
ax_box.annotate('6. Outliers', xy=(whisker_upper+0.3, 0.95), xytext=(whisker_upper+0.3,  0.95 - text_offset_y),
                    arrowprops=arrowprops, ha='center', va='top', fontsize=10)

ax_dist.set(xticks=[], xlabel="") 
plt.tight_layout()
plt.savefig("boxplot_norm_annotated.png")  
plt.show()
```
